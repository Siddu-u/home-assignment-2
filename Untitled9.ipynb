Q5 , 3. Programming ImplementationWrite Python code that:
. Accepts the confusion matrix above as input.
. Computes per-class precision and recall.
. Computes macro-averaged and micro-averaged precision and recall.
. Prints all results clearly.


```python
import numpy as np

# Confusion matrix: rows = System predictions, cols = Gold labels
# Order: [Cat, Dog, Rabbit]
cm = np.array([
    [5, 10, 5],
    [15, 20, 10],
    [0, 15, 10]
], dtype=float)

labels = ["Cat", "Dog", "Rabbit"]

# True positives are diagonal
tp = np.diag(cm)

# Predicted totals (row sums) -> for precision
pred_totals = cm.sum(axis=1)

# Gold totals (column sums) -> for recall
gold_totals = cm.sum(axis=0)

# Per-class precision and recall
precision = tp / pred_totals
recall = tp / gold_totals

# Macro averages
macro_precision = precision.mean()
macro_recall = recall.mean()

# Micro averages
total_tp = tp.sum()
total = cm.sum()

micro_precision = total_tp / total
micro_recall = total_tp / total

# Print results
print("Per-class metrics:")
for i, label in enumerate(labels):
    print(f"{label}: Precision = {precision[i]:.4f}, Recall = {recall[i]:.4f}")

print("\nMacro-averaged:")
print(f"Precision = {macro_precision:.4f}")
print(f"Recall    = {macro_recall:.4f}")

print("\nMicro-averaged:")
print(f"Precision = {micro_precision:.4f}")
print(f"Recall    = {micro_recall:.4f}")
```

Q1. Programming: Bigram Language Model Implementation (based on “Activity: I love NLP corpus” slide)


```python
# Part II - Programming
# Q1: Bigram Language Model Implementation (MLE)
# Corpus:
# <s> I love NLP </s>
# <s> I love deep learning </s>
# <s> deep learning is fun </s>

from collections import Counter

# ----------------------------
# 1) Training corpus
# ----------------------------
training_sentences = [
    ["<s>", "I", "love", "NLP", "</s>"],
    ["<s>", "I", "love", "deep", "learning", "</s>"],
    ["<s>", "deep", "learning", "is", "fun", "</s>"]
]

# ----------------------------
# 2) Compute unigram counts
# ----------------------------
unigram_counts = Counter()
for sent in training_sentences:
    unigram_counts.update(sent)

# ----------------------------
# 3) Compute bigram counts
# ----------------------------
bigram_counts = Counter()
for sent in training_sentences:
    for i in range(len(sent) - 1):
        bigram = (sent[i], sent[i + 1])
        bigram_counts[bigram] += 1

# ----------------------------
# 4) Estimate bigram probabilities using MLE
#    P(w_i | w_{i-1}) = count(w_{i-1}, w_i) / count(w_{i-1})
# ----------------------------
bigram_probs = {}
for (w1, w2), count in bigram_counts.items():
    bigram_probs[(w1, w2)] = count / unigram_counts[w1]

# ----------------------------
# 5) Function to calculate sentence probability
# ----------------------------
def sentence_probability(sentence_tokens):
    """
    Compute probability of a sentence using bigram MLE:
    P(w1, w2, ..., wn) = product of P(w_i | w_{i-1})
    Returns 0 if any bigram is unseen.
    """
    prob = 1.0
    details = []

    for i in range(len(sentence_tokens) - 1):
        bigram = (sentence_tokens[i], sentence_tokens[i + 1])

        if bigram in bigram_probs:
            p = bigram_probs[bigram]
        else:
            # Unseen bigram => probability 0 in plain MLE (no smoothing)
            p = 0.0

        details.append((bigram, p))
        prob *= p

    return prob, details

# ----------------------------
# 6) Test sentences
# ----------------------------
s1 = ["<s>", "I", "love", "NLP", "</s>"]
s2 = ["<s>", "I", "love", "deep", "learning", "</s>"]

# ----------------------------
# 7) Print counts
# ----------------------------
print("=== Unigram Counts ===")
for word, count in unigram_counts.items():
    print(f"{word}: {count}")

print("\n=== Bigram Counts ===")
for (w1, w2), count in bigram_counts.items():
    print(f"({w1} -> {w2}): {count}")

print("\n=== Bigram Probabilities (MLE) ===")
for (w1, w2), p in bigram_probs.items():
    print(f"P({w2} | {w1}) = {p:.4f}")

# ----------------------------
# 8) Compute probabilities for test sentences
# ----------------------------
p_s1, details_s1 = sentence_probability(s1)
p_s2, details_s2 = sentence_probability(s2)

print("\n=== Sentence Probability: S1 ===")
print("Sentence:", " ".join(s1))
for bg, p in details_s1:
    print(f"P({bg[1]} | {bg[0]}) = {p}")
print("Total P(S1) =", p_s1)

print("\n=== Sentence Probability: S2 ===")
print("Sentence:", " ".join(s2))
for bg, p in details_s2:
    print(f"P({bg[1]} | {bg[0]}) = {p}")
print("Total P(S2) =", p_s2)

# ----------------------------
# 9) Which sentence does the model prefer?
# ----------------------------
print("\n=== Model Preference ===")
if p_s1 > p_s2:
    print("Model prefers S1: <s> I love NLP </s>")
    print("Reason: S1 has higher probability than S2.")
elif p_s2 > p_s1:
    print("Model prefers S2: <s> I love deep learning </s>")
    print("Reason: S2 has higher probability than S1.")
else:
    print("Both sentences have equal probability.")

# Optional: print exact fractions (based on known counts)
print("\nExpected from hand calculation:")
print("P(S1) = (2/3) * (2/2) * (1/2) * (1/1) = 1/3")
print("P(S2) = (2/3) * (2/2) * (1/2) * (2/2) * (1/2) = 1/6")
```
